```python
%%R
a<-image_read("blot.png")
print(a)
a<-image_crop(a, geometry = "480x250+0+0")
```

```python
%%R
a
```

```python
%%R
a_text<-image_annotate(a, "1              2", size = 35, gravity = "North", color = "white")
a_text<-image_annotate(a_text, "Crop from blot", size = 12, gravity = "south", color = "white")

image_write(a_text, path = "crop.png", format = "png")
```

```python
%%R
blot<-ggdraw() +
draw_image("crop.png")
```

```python
%%R
C1=seq(1:20)
C2=seq(1:20)
frame<-data.frame(C1,C2)
frame
```

```python
%%R
library(tidyverse)
```

```python
%%R
graph<-frame%>%ggplot(aes(C1, C2))+theme_minimal()+annotate("segment", x = 0, xend = 0.9, y =  5.5,yend=5.5,
  colour = "black", arrow=arrow())+annotate("segment", x = 0, xend = 0.9, y =  7.5,yend=7.5,
  colour = "black", arrow=arrow())+annotate("text",x = 15, y = 15, label="protein",size=10)
```
